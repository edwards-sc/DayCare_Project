```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('C:/Users/seled/OneDrive/Desktop/DSC640/prices.xlsx')
df
```

```python
for col in df:
    print(col)
```

```python
# Group by state for MCToddler average
df_grouped_toddler = df.groupby('State_Name')['MCInfant'].mean().reset_index()
aves = df_grouped_toddler['MCInfant'].mean()
#resize
plt.figure(figsize=(10, 6))
print(aves)
#barplot for State vs MCInfant
ax = sns.barplot(data=df, x='State_Name', y='MCInfant', errorbar=None)
# create titles and labels
plt.xlabel('State Name', fontsize=12)
plt.ylabel('Median Toddler Cost', fontsize=12)
plt.xticks(rotation=90)


for i in ax.containers:
    ax.bar_label(i, )
    
```

```python
# Group by state for MCToddler average
df_grouped_toddler = df.groupby('State_Name')['MCToddler'].mean().reset_index()
ave = df_grouped_toddler['MCToddler'].mean()
#resize
plt.figure(figsize=(10, 6))
#barplot for State vs MCInfant
print(ave)
ax = sns.barplot(data=df, x='State_Name', y='MCToddler', errorbar=None)
# create titles and labels
plt.xlabel('State Name', fontsize=12)
plt.ylabel('Median Toddler Cost', fontsize=12)
plt.xticks(rotation=90)


for i in ax.containers:
    ax.bar_label(i, )
    
plt.show()
```

```python
# Group by state for MCPreschool average
df_grouped_preschool = df.groupby('State_Name')['MCPreschool'].mean().reset_index()

average = df_grouped_preschool['MCPreschool'].mean()

#resize plot
plt.figure(figsize=(10, 6))
#plot State vs MCToddler
ax = sns.barplot(data=df, x='State_Name', y='MCPreschool', errorbar=None)
# create titles and labels
print(average)

plt.xlabel('State Name', fontsize=12)
plt.ylabel('Median Preschool Cost', fontsize=12)
plt.xticks(rotation=90)


for i in ax.containers:
    ax.bar_label(i, )
    
    
plt.show()
```

```python
# create figure and axes for dual y-axes
fig, ax1 = plt.subplots(figsize=(12, 6))
# set MHI as as primary
sns.barplot(data=df, x='State_Name', y='MHI', errorbar=None, ax=ax1,color='blue')
# create label for primary y-axis
ax1.set_xlabel('State Name', fontsize=12)
ax1.set_ylabel('Median Household Income (MHI)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
plt.xticks(rotation=90) 
# Rotate x-axis
# Create secondary y-axis for Poverty Rate (PFR)
ax2 = ax1.twinx()
# Plot PFR as the secondary y-axis
sns.lineplot(data=df, x='State_Name', y='PR_F', marker='o', ax=ax2, color='orange', label='PFR')
14

# create labels for secondary y-axis
ax2.set_ylabel('Poverty Rate', fontsize=12, color='green')
ax2.tick_params(axis='y', labelcolor='green')
# place legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
# create title
plt.title('MHI and Poverty Rate by State', fontsize=16)
# show
plt.tight_layout()
plt.show()
```
